```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
m = 100
X = 2 * np.random.rand(m, 1)
Y = 4 + 3*X + np.random.rand(m, 1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y)
plt.show()
```

```python
from sklearn.preprocessing import add_dummy_feature
X_new = add_dummy_feature(X)
best_params = np.linalg.inv(X_new.T @ X_new) @ X_new.T @ Y
```

```python
best_params
```

```python
X_new1 = np.array([[0], [2]])
X_new_b = add_dummy_feature(X_new1)
y_predict = X_new_b @ best_params
y_predict
```

```python
plt.plot(X_new1, y_predict, "r-", label="Predictions")
plt.plot(X, Y, "b.")
plt.show()
```

```python
# del_theta_0 = 1/X_new.shape[0] * np.sum(hyp_fctn - Y)
# del_theta_1 = 1/X_new.shape[1] * np.sum((hyp_fctn - Y) * X_new[:, 1])
```

```python
costs = []
params_iterated = []
params = np.zeros(shape=(X_new.shape[1], 1))
# hyp_fctn = X_new @ params
for i in range(1500):
  params[0] = params[0] - 0.022 * 1/X_new.shape[0] * np.sum(X_new @ params - Y)
  params[1] = params[1] - 0.022 * 1/X_new.shape[1] * np.sum(X.T @ (X_new @ params - Y))
  if i%200 == 0:
    print(f"\t pr0:{params[0]}\t pr1:{params[1]}")
  params_iterated.append(params.copy())
  cost_fctn = 1/(2*X_new.shape[0]) * np.sum((X_new @ params - Y)**2)
  costs.append(cost_fctn)
```

```python
mincst = min(costs)
```

```python
indx = 0
for i, v in enumerate(costs):
  if v==mincst:
    indx=i
    break;
indx
```
